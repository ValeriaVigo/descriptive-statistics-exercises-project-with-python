# Descriptive statistics problems

### Exercise 1

We will use NumPy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
import numpy as np
from scipy.stats import skew
from scipy.stats import kurtosis

# Generar un array de 100 elementos siguiendo una distribución normal
array_normal = np.random.normal(size=100)

print("array normal:", array_normal)

# Generar un array de 100 elementos siguiendo una distribución chi-cuadrado con 3 grados de libertad
array_chi_cuadrado = np.random.chisquare(df=3, size=100)

print("array chi cuadrado:", array_chi_cuadrado)

# Calcular las principales métricas y medidas estadísticas que mejor describen los dos vectores
# Para el array normal
normal_mean = np.mean(array_normal)
normal_std = np.std(array_normal)
normal_var = np.var(array_normal)
normal_median = np.median(array_normal)
skew_normal = skew(array_normal)
kurt_normal = kurtosis(array_normal)

# Para el array de chi-cuadrado
chi_mean = np.mean(array_chi_cuadrado)
chi_std = np.std(array_chi_cuadrado)
chi_var = np.var(array_chi_cuadrado)
chi_median = np.median(array_chi_cuadrado)
skew_chi = skew(array_chi_cuadrado)
kurt_chi = kurtosis(array_chi_cuadrado)

# Imprimir los resultados
print("\nPara el array normal:")
print("Media:", normal_mean)
print("Desviación estándar:", normal_std)
print("Varianza:", normal_var)
print("Mediana:", normal_median)
print("Skewness:", skew_normal)
print("Kurtosis:", kurt_normal)
print("\nPara el array de chi-cuadrado:")
print("Media:", chi_mean)
print("Desviación estándar:", chi_std)
print("Varianza:", chi_var)
print("Mediana:", chi_median)
print("Skewness:", skew_chi)
print("Kurtosis:", kurt_chi)
````

### Exercise 2

Write a Python program to calculate the standard deviation of the following data:

```py
data = [4, 2, 5, 8, 6]
```

````python
import math
import sys

# Datos
data = [4, 2, 5, 8, 6]

#Media
mean = sum(data) / len(data)

# Suma de Cuadrado
sum_of_sq_diff = sum((x - mean) ** 2 for x in data)

# Varianza
variance = sum_of_sq_diff / len(data)

# Desviación estándar
standard_deviation = math.sqrt(variance)

print("La desviación estándar es:", standard_deviation)



````
